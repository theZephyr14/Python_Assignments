```python

import pandas as pd
import os
os.getcwd()

path = r'c:\\Users\\Kevin\\Desktop\\ESADE_Classes\\Python\\session6\\titanic\\train_and_test2.csv'

df = pd.read_csv(path)
df
```

```python

df["Sex"] = df["Sex"].fillna(1)
df["2urvived"] = df["2urvived"].fillna(0)

survived = df[df["2urvived"] == 1]
gender_counts = survived["Sex"].value_counts()

male_count = (survived["Sex"] == 1).sum()
female_count = (survived["Sex"] == 0).sum()

ratio = male_count / female_count if female_count > 0 else float('inf')
print(f"Ratio of males to females among survivors: {ratio}")
```

```python
# mean_fare = df2['Fare'].mean()

#Class
mean_fare = df['Fare'].mean()

upper_class = df[df['Fare'] >= mean_fare]
lower_class = df[df['Fare'] < mean_fare]

passenger_tot = df.groupby(df['Fare'] >= mean_fare).size()
upper_class_count = passenger_tot.get(True, 0)
lower_class_count = passenger_tot.get(False, 0)

upper_class_survived = upper_class['2urvived'].sum() 
lower_class_survived = lower_class['2urvived'].sum()

upper_class_survival_percentage = (upper_class_survived / upper_class_count) * 100
lower_class_survival_percentage = (lower_class_survived / lower_class_count) * 100

print("The percentage of upper class that survived is: ", upper_class_survival_percentage)
print("The percentage of lower class that survived is: ", lower_class_survival_percentage)

#Gender
male_sur = df[(df["Sex"] == 1) & (df["2urvived"] == 1)]
female_sur = df[(df["Sex"] == 0) & (df["2urvived"] == 1)]

tot = len(male_sur) + len(female_sur)

if tot > 0:
    male_per = (len(male_sur) / tot) * 100
    female_per = (len(female_sur) / tot) * 100
else:
    male_per = female_per = 0

print("Males survived percentage:", male_per)
print("Females survived percentage:", female_per)
```
